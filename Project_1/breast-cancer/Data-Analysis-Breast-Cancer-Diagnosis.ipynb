```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import genfromtxt

cancer_data = genfromtxt('breast-cancer-wisconsin.csv', delimiter=',')
```

```python
shape = np.shape(cancer_data)
s = shape[0]
```

```python
np_hist = np.zeros(s)
for x in range(1, s):
    a = cancer_data[x,1]
    np_hist[x] = a
hist,bin_edges = np.histogram(np_hist)

plt.figure(figsize=[8,5])

plt.bar(bin_edges[:-1], hist, width = 0.4, color='#0504aa',alpha=0.8)
plt.xlim(min(bin_edges), max(bin_edges))
plt.grid(axis='y', alpha=0.85)
plt.xlabel('Value',fontsize=12)
plt.ylabel('Frequency',fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel('Frequency',fontsize=12)
plt.title('Distribution of Clump Thickness',fontsize=12)
plt.show()
```

```python
b = 0;
m = 0;
#Calculate the total number of benign and malignant instances in the data set
for x in range(0,s):
    if cancer_data[x,10] == 2:
        b = b + 1
        
    elif cancer_data[x,10] == 4:
        m = m + 1
        
#Pie chart representation of the binary output
labels = 'benign', 'malignant'
b = b*100/(b+m)
m = m*100/(b+m)
sizes = [b, m]
#Separates both slices
explode = (0.025, 0.025) 

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
ax1.axis('equal')  # Makes the pie chart look circle
plt.show()
```
